```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("Mall_Customers.csv")
x = ds.iloc[:,[3,4]].values
```

```python
# Elbow Methode
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    k = KMeans(n_clusters=i, init='k-means++', random_state=42)
    k.fit(x)
    wcss.append(k.inertia_)
plt.plot(range(1,11), wcss)
plt.title("Elbow Methode")
plt.xlabel("number of cluster")
plt.ylabel("Wcss")
plt.show()    
```

```python
# Training the model on the whole dataset
k = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_k = k.fit_predict(x)
```

```python
# Visualizaing 
plt.scatter(x[y_k == 0, 0], x[y_k == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_k == 1, 0], x[y_k == 1, 1], s = 100, c = 'Orange', label = 'Cluster 2')
plt.scatter(x[y_k == 2, 0], x[y_k == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_k == 3, 0], x[y_k == 3, 1], s = 100, c = 'pink', label = 'Cluster 4')
plt.scatter(x[y_k == 4, 0], x[y_k == 4, 1], s = 100, c = 'purple', label = 'Cluster 5')
plt.scatter(k.cluster_centers_[:,0], k.cluster_centers_[:,1], s=300, c = 'yellow', label = 'Centroide')
plt.title("Customer cluster")
plt.xlabel("Annual Income")
plt.ylabel("Spending score(1-100)")
plt.legend()
plt.show()
```
