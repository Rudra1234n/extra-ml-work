## K-Means Clustering 

Importing Lib


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

importing the dataset 

```python
ds = pd.read_csv("Mall_Customers.csv")
x = ds.iloc[:,[3,4]].values
```

Using the elbow methode to find the optimal number clusters


```python
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    k = KMeans(n_clusters=i, init='k-means++', random_state=42)
    k.fit(x)
    wcss.append(k.inertia_)
plt.plot(range(1,11), wcss)
plt.title("Elbow methode")
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
plt.show()
    
```

Training the model on dataset

```python
k = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_k = k.fit_predict(x)
```

Visualizing the clusters

```python
plt.scatter(x[y_k == 0, 0], x[y_k==0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_k == 1, 0], x[y_k==1, 1], s = 100, c = 'orange', label = 'Cluster 2')
plt.scatter(x[y_k == 2, 0], x[y_k==2, 1], s = 100, c = 'pink', label = 'Cluster 3')
plt.scatter(x[y_k == 3, 0], x[y_k==3, 1], s = 100, c = 'blue', label = 'Cluster 4')
plt.scatter(x[y_k == 4, 0], x[y_k==4, 1], s = 100, c = 'green', label = 'Cluster 5')
# making of centroids
plt.scatter(k.cluster_centers_[:,0], k.cluster_centers_[:,1], s = 300, c = 'yellow', label = 'Centroids')
plt.title("Clusters of Customers")
plt.xlabel("Anual income")
plt.ylabel("Pending score")
plt.legend()
plt.show()
```
