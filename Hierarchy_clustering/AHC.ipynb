## Hierarchical Clustring

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import scipy.cluster.hierarchy as sch
```

```python
ds = pd.read_csv("Mall_Customers.csv")
x = ds.iloc[:,[3,4]].values
```

```python
print(x)
```

Building the dendogram

```python
d = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title("Denedrogram")
plt.xlabel("Obsevation Point")
plt.ylabel("Euclidean Distance")
plt.show()
```

Training the HC model on dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(x)
```

Visualizing the model

```python
plt.scatter(x[y_hc==0, 0], x[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_hc==1, 0], x[y_hc == 1, 1], s = 100, c = 'yellow', label = 'Cluster 2')
plt.scatter(x[y_hc==2, 0], x[y_hc == 2, 1], s = 100, c = 'pink', label = 'Cluste 3')
plt.scatter(x[y_hc==3, 0], x[y_hc == 3, 1], s = 100, c = 'blue', label = 'Cluster 4')
plt.scatter(x[y_hc==4, 0], x[y_hc == 4, 1], s = 100, c = 'green', label = 'Cluster 5')
plt.title("Hierarchcal Clustering")
plt.xlabel("Annual Income($)")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
