## Simple Linear Regression 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv('Salary_Data.csv')
x = ds.iloc[:,:-1].values
y = ds.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

Training the model on simple linear regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
# x_train = Independent Variable
# y_train = Dependent Variable
```

Predicting the model

```python
y_pred = lr.predict(x_test)
```

Visualizing on the training set

```python
plt.scatter(x_train,y_train, color='green')
plt.plot(x_train, lr.predict(x_train),  color='yellow') # Best fit line code
plt.title("Salaries V/S Years")
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()
```

Visulalizting on test set

```python
plt.scatter(x_train,y_train, color='green')
plt.plot(x_train, lr.predict(x_train),  color='yellow') # Best fit line code
plt.title("Salaries V/S Years")
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()
```
